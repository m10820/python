```python
#kernel density estimation plots
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset,height=0.25)
# every tick is a spot in that dataset 

plt.ylim(0,1)
```

```python
plt.hist(dataset,alpha=0.3)
sns.rugplot(dataset,height=0.05)
#tick is sort of a tally mark 
```

```python
#the kernal density point , represent each of these tick marks
#with a gaussian basis function
```

```python
#creata rugplot
sns.rugplot(dataset,height=0.25)
#set our x min max value
x_min = dataset.min() - 2
x_max = dataset.max() + 2
#create x axis,value is 100 linearly spaced pnts btw the min nd max
x_axis = np.linspace(x_min,x_max,100)

#for setting bandwith
bandwidth =(4*dataset.std()**5) / (3*len(dataset)) ** 0.2

kernel_list = []
#for every datapt in dataset we create and plot that kernel
for data_point in dataset:
    #create a kernel for each point nd append it to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list,axis=0)

fig = plt.plot(x_axis,sum_of_kde,color='indianred')

sns.rugplot(dataset)

plt.yticks([])
plt.suptitle('sum of basis functions')

#now we have kernel density estimation plot by adding up
#all gaussian functs for each tick marks
```
