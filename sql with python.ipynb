```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("sakila (1).db")
```

```python
# Set SQL query as a comment
sql_query = ''' SELECT * FROM customer '''

# Use pandas to pass sql query using connection form SQLite3
df = pd.read_sql(sql_query, con)

# Show the resulting DataFrame
df.head()
```

```python
def sql_to_df(query):
    df = pd.read_sql(query,con)
    return df
```

```python
q = ''' SELECT first_name,last_name FROM customer; '''
sql_to_df(q).head()
```

```python
#DISTINCT in sql means unique values
```

```python
q = ''' SELECT DISTINCT(country_id) FROM city '''
```

```python
sql_to_df(q)
```

```python
q = ''' SELECT *
        FROM customer
        WHERE first_name = 'MARY' '''
```

```python
sql_to_df(q)
```

```python
q = ''' SELECT *
        FROM customer
        WHERE store_id = 1 '''
```

```python
sql_to_df(q)
```

```python
#for choosing a specific cell use'Mary' to get cell with name mary
```

```python
q = ''' SELECT AVG(customer_id)
        FROM customer; '''
```

```python
sql_to_df(q)
```

```python
## % for 0 or more characters
## _ for a single character
## [] for sets and ranges of character to match
```

```python
q = ''' SELECT *
        FROM customer
        WHERE first_name LIKE 'M%' ; '''
#every first name starting from M

sql_to_df(q)
```

```python
q = ''' SELECT *
        FROM customer
        WHERE last_name LIKE '_ING' ; '''
# every lastname end with ing having only 1 character before ing
sql_to_df(q)
```

```python
q = ''' SELECT *
        FROM customer
        WHERE first_name LIKE 'MA%' ; '''

sql_to_df(q)
```

```python
#select customer whose name begin with either a or b
#character list
q = ''' SELECT *
        FROM customer
        WHERE first_name GLOB '[AB]*' ; '''

sql_to_df(q)
```

```python
q = ''' SELECT *
        FROM customer
        ORDER BY last_name DESC ; '''

sql_to_df(q)
#alphabetic order
```

```python
#group result by one or more columns
#count number of customers per store
q = ''' SELECT store_id , COUNT(customer_id)
        FROM customer
        GROUP BY store_id; '''

sql_to_df(q)
```

```python
q = ''' SELECT store_id , COUNT(customer_id)
        FROM customer; '''

sql_to_df(q)
```
